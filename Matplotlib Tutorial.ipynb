### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5) # start at 0 count up by 0.5 to 4.5 but not include 4.5
#print(x2) # [0.  0.5 1.  1.5 2.  2.5 3.  3.5 4. ]

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2') # from beginning to 5th index

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--') # from 5th index to end

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

# Display a Bar Chart
bars = plt.bar(labels, values) # can also use barh for horizontal bar chart

# Different Patterns for Bar Columns
patterns = ['/', '////', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(5,3), dpi=100)

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

# Access the data with pandas via the column name
plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea') # This is how you have to reference columns with a space in the name
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

# Another Way to plot many values!
# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::3].tolist()+[2011]) # This will only show every 3rd year and the last year

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend() # With this you don't need to specify label in the plot function, this is because in the csv file the first row is the name of the columns

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head()
```

### Histogram

```python
plt.figure(figsize=(8,5) , dpi=100)

# Show a Histogram, the bins parameter specifies the number of bins we want
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

```python
# Get the count of players with preferred foot, left or right
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

# Create a Pie Chart, the autopct parameter shows the percentage of each slice
# The first parameter is the values for each slice
plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart #2

```python
plt.figure(figsize=(8,6), dpi=100)

plt.style.use('ggplot') # This is a style that makes the plot look better

# In the csv file the weight is in lbs, so we need to remove the lbs and convert to int
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# Get the weight of players and divide them into 5 groups
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4) # This will separate the 1st, 2nd and last slice and make the smaller slices more visible

plt.title('Weight of Professional Soccer Players (lbs)')

# autopct will show the percentage of each slice, pctdistance will move the percentage away from the center
plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

# Get lists of the overall ratings for each team
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

# Create a Box Plot for the 3 teams
# Automatically includes the median, 1st and 3rd quartile and min and max values
#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})
# Set patch_artist=True enable color customization

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')
    
plt.show()
```
